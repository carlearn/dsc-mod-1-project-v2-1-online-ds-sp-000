### The Numbers

```python
df_tn = pd.read_csv('zippedData/tn.movie_budgets.csv.gz')
print(df_tn.shape)
df_tn.head()
```

```python
df_tn['release_date'] = pd.to_datetime(df_tn['release_date'])
df_tn['year'] = pd.DatetimeIndex(df_tn['release_date']).year
```

```python
df_tn = df_tn[df_tn['year'] >= 2010]
```

```python
df_tn_new = df_tn.copy()
```

```python
cols_to_clean = ['production_budget', 'domestic_gross', 'worldwide_gross']
for col in cols_to_clean:
    df_tn_new[col] = df_tn_new[col].map(lambda x: x.replace('$',''))
```

```python
for col in cols_to_clean:
    df_tn_new[col] = [x.replace(',', '') if ',' in x else x for x in df_tn_new[col]]
```

```python
for col in cols_to_clean:
    df_tn_new[col] = df_tn_new[col].astype('int')
```

```python
df_tn_new
```

```python
df_tn_new['profit'] = df_tn_new['worldwide_gross'] - df_tn_new['production_budget']
```

```python
df_tn_new.drop('id', axis=1)
```

```python
df_profit = df_tn_new[['movie','release_date','production_budget','worldwide_gross','profit']].sort_values(by='profit', ascending=False)
```

```python
df_profit
```

# Thoughs on loading and previewing the datasets

There's a problem with joining datasets from multiple sources, as formats are almost never standardized across the sources.  For example, the movie "Star Wars: The Force Awakens" has different title format in IMDb, TMDb and TN datasets. Based on the discussion with my instructor Jeff, he provided me with some solutions. 

When coming across such problem, there are three solutions:

- Are there many edge cases like this? If there’s aren’t many edge cases, you can manually clean the titles;
- If there are a bunch of edge cases, another option would be to grab partial titles (maybe just grab the first couple of words from the title) and join on the partial title and the year the film was released;
- If the above two solutions do not work, you can try APIs to see if there's a unique key that you can join on.

I tried to use partial title and year the film was released, but still missed some movie titles with special symbols. Except some common apostrophes such as comma, colon and dash, some apostrophes in movie titles are not easily to find out. Most of the titles incurred with unmatched problems are those popular series movies with substantial revenues. For example, Fast & Furious.

Therefore, my final solution is to get more complete datasets from TMDb API where I can get movie title, release_year, budget, revenue, popularity and imdb_id. Accordingly, I can use the unique imdb_id to link the datasets provided to get more information of directors and actors.
